### Task 1: Preprocessing
1. Convert text to lowercase
2. Tokenization
3. Remove punctuation
4. Optional: Remove stopwords (and, a, the)
5. Optional: Lemmatization

```python
import pandas as pd
import string

# Load dataset
DATA_PATH = "data/amazon_reviews.csv"
OUTPUT_PATH = "data/cleaned_reviews.csv"

df = pd.read_csv(DATA_PATH)

print("Loaded rows:", len(df))
print("Columns:", df.columns.tolist())
print("Missing values per column:\n", df.isnull().sum())

# Basic preprocessing steps:
# 1. lowercase
# 2. remove punctuation
# 3. tokenize by whitespace
# 4. remove simple stopwords
# Optional lemmatization is not applied here because only pandas and numpy are installed.
STOPWORDS = {"and", "a", "the", "is", "it", "this", "that", "was", "for", "to", "of", "in", "on", "with", "as", "its", "at", "an"}


def clean_text(text: str) -> str:
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = text.split()
    tokens = [token for token in tokens if token not in STOPWORDS]
    return " ".join(tokens)

for col in ["Review Title", "Review Description"]:
    if col in df.columns:
        df[f"{col} Cleaned"] = df[col].apply(clean_text)

cleaned_df = df[["Star Rating", "Review Title Cleaned", "Review Description Cleaned"]].copy()
cleaned_df.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

print("Saved cleaned dataset to:", OUTPUT_PATH)
print("Cleaned dataset shape:", cleaned_df.shape)
print(cleaned_df.head(3).to_dict(orient="records"))
```
